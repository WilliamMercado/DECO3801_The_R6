```python
from sd_backbone import StableDiffusionBackBone
from PIL import Image

bb = StableDiffusionBackBone()
```

```python
bb.reset()

ref_img_path = 'C:/Users/q3357/Pictures/GQl6fs_akAAMzef.jpg'
ref_img = Image.open(r'C:/Users/q3357/Pictures/GQl6fs_akAAMzef.jpg')

color_grid_path = 'C:/Users/q3357/Pictures/redbull.png'
color_grid = Image.open(r'C:/Users/q3357/Pictures/redbull.png')

import matplotlib.pyplot as plt

ls = [ref_img, color_grid]

fig, axes = plt.subplots(1, len(ls), figsize=(15, 5))
for ax, img in zip(axes, ls):
    ax.imshow(img)
    ax.axis('off')

plt.show()
```

```python
bb.add_control_unit(unit_num=0, image_path=ref_img_path, module='canny')
bb.add_control_unit(unit_num=1, image_path=color_grid_path, module='shuffle', intensity='high')
```

```python
out = bb.txt2img('1 anime girl')
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, len(out), figsize=(15, 5))
for ax, img in zip(axes, out):
    ax.imshow(img)
    ax.axis('off')

plt.show()
```
